## Question 2

Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python

# import libraries

import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load the JSON data from the file and read using 'r'
with open('lyrics_object.json', 'r') as json_file:
    lyrics_data = json.load(json_file)

# Load spacy model and create spacy pipeline
nlp = spacy.load('en_core_web_sm') 
nlp.add_pipe('spacytextblob')

# Get the lyrics from the dictionary using .get function
lyrics = lyrics_data.get('lyrics')

# Process the lyrics
model_doc = nlp(lyrics)

# Perform sentiment analysis and get the polarity score
sentiment = model_doc._.blob.polarity

# Print the lyrics and the polarity score
print('Lyrics:\n') # add \n to create a new line after Lyrics: prints
print(lyrics) # print lyrics from lyrics variable
print()
print()
print()
print('\nPolarity score:', sentiment) # use \n before print Polarity score to create new line before print statement

# Create a conditional statement to create variable comment giving sentiment comment based on score, closer to 1 is positive, closer to -1 is negative
if sentiment > 0:
    comment = 'positive'
elif sentiment < 0:
    comment = 'negative'
else:
    comment = 'neutral'

print() # add an empty line
print('The lyrics have a', comment, 'connotation.') # create print statement for sentiment connotation 

```

## Question 3

Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
# import libraries

import json
import requests

def get_lyrics(artist, song, file):
    # Create the API URL based on artist and song
    url = f'https://api.lyrics.ovh/v1/{artist}/{song}' 
    
    # Make a request to the API
    response = requests.get(url)
        
    # If the request was successful with code being 200 parse the JSON response
    if response.status_code == 200:
        result = response.json()              
        # Write the lyrics to the file
        with open(file, 'w') as json_file:
            json.dump(result, json_file, indent=4)
                
        print(f"Lyrics for '{song}' by {artist} saved to {file}.")
    # If request not successful give error comment   
    else:
        print(f"Could not get lyrics for '{song}' by {artist} due to error code: {response.status_code}")


# Test the function with four songs saved to files

# Song 1 "Speed of Sound" by Coldplay
get_lyrics('Coldplay', 'Speed of Sound', 'speed_of_sound.json')

# Song 2 "Midnight City" by M83
get_lyrics('M83', 'Midnight City', 'midnight_city.json')

# Song 3 "Radioactive" by Imagine Dragones
get_lyrics('Imagine Dragons', 'Radioactive', 'radioactive.json')

# Song 4 "In the End" by Linkin Park
get_lyrics('Linkin Park', 'In the End', 'in_the_end.json')
```

## Question 4

Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load spacy model and SpacyTextBlob
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")

def sentiment_analysis(file, song):
    # Load the lyrics from JSON
    with open(file, 'r') as json_file:
        lyrics_data = json.load(json_file)
    
    # Get the lyrics
    lyrics = lyrics_data.get('lyrics', '')
    
    # Sentiment analysis
    doc = nlp(lyrics)
    sentiment = doc._.blob.polarity
    
    # Print song name and its polarity score as sentiment
    print(f"Polarity Score for lyrics in '{song}': {sentiment}")
    return sentiment

# Test the function with 3 JSON files
sentiment_analysis("speed_of_sound.json", "Speed of Sound")
sentiment_analysis("midnight_city.json", "Midnight City")
sentiment_analysis("radioactive.json", "Radioactive")
sentiment_analysis("in_the_end.json", "In the End")
```
